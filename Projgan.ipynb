```python
# Part 0: Setup and Global Configurations
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.utils.class_weight import compute_class_weight
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import glob
import os

# For reproducibility
SEED = 42
np.random.seed(SEED)
tf.random.set_seed(SEED)

# Define global constants
DATA_PATH = "data/" # Define the path to the data folder
SEQUENCE_LENGTH = 1000  # Sequence length in data points
LATENT_DIM_AE = 50      # Latent dimension for the Autoencoder
LATENT_DIM_GAN = 100    # Latent dimension for the GAN
GAN_TRAIN_EPOCHS = 1000 # Example epochs for GAN
GAN_BATCH_SIZE = 32     # Example batch size for GAN
AE_TRAIN_EPOCHS = 100   # Example epochs for Autoencoder
CLASSIFIER_TRAIN_EPOCHS = 100 # Example epochs for Classifier
CLASSIFIER_BATCH_SIZE = 64    # Example batch size for Classifier

# --- Part 1: Data Loading and Preparation ---
print("--- Part 1: Data Loading and Preparation ---")

def load_and_prepare_data(data_path_folder):
    """
    Loads data from specific CSV files with 3 categories: quiet, vehicle, human.
    This function loads only the first column of each CSV and concatenates
    all data points into a single stream per class before combining.
    """
    file_mapping = {
        'car_nothing(AVI).csv': 'quiet',
        'carnew(AVI).csv': 'vehicle',
        'human_nothing(AVI).csv': 'quiet',
        'human(AVI).csv': 'human'
    }
    label_encoding = {'quiet': 0, 'vehicle': 1, 'human': 2}
    all_data = []
    all_labels = []

    print("Starting data loading...")
    for filename, activity_type in file_mapping.items():
        filepath = os.path.join(data_path_folder, filename)
        if not os.path.exists(filepath):
            print(f"Error: File not found at {filepath}. Skipping.")
            continue
        try:
            df = pd.read_csv(filepath, header=None)
            if not df.empty and df.shape[1] > 0:
                data = df.iloc[:, 0].values
                print(f"Read file {filename}: {df.shape[0]} samples -> {activity_type}")
                label_code = label_encoding[activity_type]
                all_data.extend(data)
                all_labels.extend([label_code] * len(data))
            else:
                print(f"Warning: File {filename} is empty or has no data columns. Skipping.")
        except Exception as e:
            print(f"Error reading {filename}: {e}")

    print("Data loading finished.")
    all_data_np = np.array(all_data)
    all_labels_np = np.array(all_labels)

    print(f"Total data points loaded: {len(all_data_np)}")
    print(f"Total labels loaded: {len(all_labels_np)}")
    if len(all_labels_np) > 0:
        print(f"Unique categories loaded: {np.unique(all_labels_np)} - {list(label_encoding.keys())}")
        print(f"Label distribution (counts): {np.bincount(all_labels_np)}")
    else:
        print("No labels loaded.")
    return all_data_np, all_labels_np

X_raw, y_raw = load_and_prepare_data(DATA_PATH)

if len(X_raw) == 0:
    print("No data loaded. Exiting.")
    exit()
else:
    print("Raw data loaded successfully.")

# --- Part 2: Sequence Creation, Normalization, and Data Splitting ---
print("\n--- Part 2: Sequence Creation, Normalization, and Data Splitting ---")

def create_sequences(data, labels, sequence_length):
    """
    Creates fixed-size sequences from the concatenated data stream.
    Sequences are created by chunking the data belonging to each label separately.
    Trailing data points for each class that don't form a full sequence are discarded.
    """
    sequences = []
    sequence_labels = []
    unique_labels = np.unique(labels)
    print(f"\nCreating sequences of length {sequence_length}...")

    for label in unique_labels:
        label_indices = np.where(labels == label)[0]
        label_data = data[label_indices]
        num_sequences_for_label = len(label_data) // sequence_length
        print(f"Processing label {label} (count: {len(label_data)}): Can create {num_sequences_for_label} sequences.")
        for i in range(num_sequences_for_label):
            start_idx = i * sequence_length
            end_idx = start_idx + sequence_length
            sequences.append(label_data[start_idx:end_idx])
            sequence_labels.append(label)

    print(f"Sequence creation finished. Total sequences created: {len(sequences)}")
    return np.array(sequences), np.array(sequence_labels)

X_sequences, y_sequences = create_sequences(X_raw, y_raw, SEQUENCE_LENGTH)

if len(X_sequences) == 0:
    print("No sequences were created. Check data loading and sequence_length. Exiting.")
    exit()

print(f"Created {len(X_sequences)} sequences with length {SEQUENCE_LENGTH}.")
print(f"Sequence data shape: {X_sequences.shape}")
print(f"Sequence labels shape: {y_sequences.shape}")
unique_seq_labels, seq_label_counts = np.unique(y_sequences, return_counts=True)
print(f"Distribution of sequence labels: {dict(zip(unique_seq_labels, seq_label_counts))}")

# Data Splitting (before normalization to fit scaler only on train)
X_train_seq, X_temp_seq, y_train, y_temp = train_test_split(
    X_sequences, y_sequences, test_size=0.3, random_state=SEED, stratify=y_sequences
)
X_val_seq, X_test_seq, y_val, y_test = train_test_split(
    X_temp_seq, y_temp, test_size=0.5, random_state=SEED, stratify=y_temp
)

print(f"\nData split into:")
print(f"  Training set: {X_train_seq.shape[0]} sequences")
print(f"  Validation set: {X_val_seq.shape[0]} sequences")
print(f"  Testing set: {X_test_seq.shape[0]} sequences")
print("Label distribution in splits:")
print(f"  Train: {np.bincount(y_train)}")
print(f"  Validation: {np.bincount(y_val)}")
print(f"  Test: {np.bincount(y_test)}")

# Normalization (Fit on train, transform train, val, test)
scaler = StandardScaler()

# Reshape for scaler: (n_samples * sequence_length, 1)
X_train_flat = X_train_seq.reshape(-1, 1)
X_val_flat = X_val_seq.reshape(-1, 1)
X_test_flat = X_test_seq.reshape(-1, 1)

# Fit scaler ONLY on training data
scaler.fit(X_train_flat)

# Transform all sets
X_train_normalized_flat = scaler.transform(X_train_flat)
X_val_normalized_flat = scaler.transform(X_val_flat)
X_test_normalized_flat = scaler.transform(X_test_flat)

# Reshape back to (n_samples, sequence_length)
X_train_normalized = X_train_normalized_flat.reshape(X_train_seq.shape)
X_val_normalized = X_val_normalized_flat.reshape(X_val_seq.shape)
X_test_normalized = X_test_normalized_flat.reshape(X_test_seq.shape)
print("\nData normalized using StandardScaler (fitted on training data only).")

# Reshape for Keras Conv1D (add channel dimension)
X_train_reshaped = X_train_normalized.reshape(-1, SEQUENCE_LENGTH, 1)
X_val_reshaped = X_val_normalized.reshape(-1, SEQUENCE_LENGTH, 1)
X_test_reshaped = X_test_normalized.reshape(-1, SEQUENCE_LENGTH, 1)

print("\nData reshaped for Keras (added channel dimension).")
print(f"Final Training shape: {X_train_reshaped.shape}")
print(f"Final Validation shape: {X_val_reshaped.shape}")
print(f"Final Testing shape: {X_test_reshaped.shape}")


# --- Part 3: Autoencoder Model Definition, Compilation, and Training ---
print("\n--- Part 3: Autoencoder Model ---")

def build_autoencoder(input_shape, encoding_dim):
    input_layer = keras.Input(shape=input_shape)
    # Encoder
    x = layers.Conv1D(32, 3, activation='relu', padding='same')(input_layer)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2, padding='same')(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Conv1D(16, 3, activation='relu', padding='same')(x)
    x = layers.BatchNormalization()(x)
    x = layers.MaxPooling1D(2, padding='same')(x)
    x = layers.Dropout(0.2)(x)
    x = layers.Conv1D(8, 3, activation='relu', padding='same')(x)
    pooled_output = layers.GlobalMaxPooling1D()(x)
    encoded = layers.Dense(encoding_dim, activation='relu',
                           kernel_regularizer=keras.regularizers.l2(0.001),
                           name='encoded')(pooled_output)
    # Decoder
    decoder_dense_size = 8 * (input_shape[0] // 4)
    x = layers.Dense(decoder_dense_size, activation='relu')(encoded)
    x = layers.Reshape((input_shape[0] // 4, 8))(x)
    x = layers.Conv1D(8, 3, activation='relu', padding='same')(x)
    x = layers.UpSampling1D(2)(x)
    x = layers.Conv1D(16, 3, activation='relu', padding='same')(x)
    x = layers.UpSampling1D(2)(x)
    decoded = layers.Conv1D(1, 3, activation='linear', padding='same')(x)

    autoencoder_model = keras.Model(input_layer, decoded, name='autoencoder')
    encoder_model = keras.Model(input_layer, encoded, name='encoder')
    return autoencoder_model, encoder_model

autoencoder, encoder = build_autoencoder(
    input_shape=(SEQUENCE_LENGTH, 1),
    encoding_dim=LATENT_DIM_AE
)

optimizer_ae = keras.optimizers.Adam(learning_rate=0.001)
autoencoder.compile(optimizer=optimizer_ae, loss='mse', metrics=['mae'])

ae_callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=5, min_lr=1e-6)
]

print("\nAutoencoder Architecture:")
autoencoder.summary()
# print("\nEncoder Architecture (part of the Autoencoder):")
# encoder.summary() # Optional

print("\nStarting Autoencoder training...")
history_ae = autoencoder.fit(
    X_train_reshaped, X_train_reshaped, # Autoencoder reconstructs its input
    epochs=AE_TRAIN_EPOCHS,
    batch_size=CLASSIFIER_BATCH_SIZE, # Can use same batch size or different
    shuffle=True,
    validation_data=(X_val_reshaped, X_val_reshaped),
    callbacks=ae_callbacks,
    verbose=1
)
print("Autoencoder training finished.")

# Optional: Plot Autoencoder training history
plt.figure(figsize=(10, 4))
plt.plot(history_ae.history['loss'], label='AE Train Loss')
plt.plot(history_ae.history['val_loss'], label='AE Validation Loss')
plt.title('Autoencoder Training Loss')
plt.xlabel('Epoch')
plt.ylabel('Loss (MSE)')
plt.legend()
plt.grid(True)
plt.show()

# --- Part 4: Generative Adversarial Network (GAN) Model ---
print("\n--- Part 4: Generative Adversarial Network (GAN) ---")
# Note on GAN data: GANs often benefit from data scaled to a specific range like [-1, 1] (e.g., using Tanh in generator).
# StandardScaler normalizes to mean 0, std 1. If Tanh is used in generator,
# consider applying MinMaxScaler to [-1, 1] on X_train_reshaped specifically for GAN training.
# For simplicity, we'll use the StandardScaler output here.

def build_generator_gan(latent_dim, output_shape_gan):
    model = keras.Sequential(name='generator')
    model.add(layers.Input(shape=(latent_dim,)))
    model.add(layers.Dense(128, activation='relu'))
    model.add(layers.Dense(256, activation='relu'))
    # Calculate required dense size to reshape correctly
    # For Conv1D output_shape_gan is (SEQUENCE_LENGTH, 1)
    required_dense_size = np.prod(output_shape_gan)
    model.add(layers.Dense(required_dense_size, activation='tanh')) # Tanh outputs [-1, 1]
    model.add(layers.Reshape(output_shape_gan))
    return model

def build_discriminator_gan(input_shape_gan):
    model = keras.Sequential(name='discriminator')
    model.add(layers.Input(shape=input_shape_gan))
    model.add(layers.Conv1D(64, 3, strides=2, padding='same'))
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.25))
    model.add(layers.Conv1D(128, 3, strides=2, padding='same'))
    model.add(layers.LeakyReLU(alpha=0.2))
    model.add(layers.Dropout(0.25))
    model.add(layers.GlobalMaxPooling1D())
    model.add(layers.Dense(1, activation='sigmoid')) # Probability (real/fake)
    return model

def build_combined_gan(gen, disc):
    disc.trainable = False
    gan_input_combined = keras.Input(shape=(LATENT_DIM_GAN,))
    generated_signal_combined = gen(gan_input_combined)
    gan_output_combined = disc(generated_signal_combined)
    gan_combined = keras.Model(gan_input_combined, gan_output_combined, name='gan')
    return gan_combined

generator_gan = build_generator_gan(LATENT_DIM_GAN, (SEQUENCE_LENGTH, 1))
discriminator_gan = build_discriminator_gan((SEQUENCE_LENGTH, 1))
gan_combined_model = build_combined_gan(generator_gan, discriminator_gan)

# GAN Optimizers (often specific settings are used)
opt_discriminator_gan = keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)
opt_gan_combined = keras.optimizers.Adam(learning_rate=0.0002, beta_1=0.5)

discriminator_gan.compile(optimizer=opt_discriminator_gan, loss='binary_crossentropy', metrics=['accuracy'])
gan_combined_model.compile(optimizer=opt_gan_combined, loss='binary_crossentropy')

print("\nGenerator Architecture:")
generator_gan.summary()
print("\nDiscriminator Architecture:")
discriminator_gan.summary()

def train_gan_loop(gen, disc, combined_gan, data_for_gan, epochs, batch_size_gan, latent_dim_gan):
    num_batches = data_for_gan.shape[0] // batch_size_gan
    if num_batches == 0:
        print("Not enough data to form a batch for GAN training. Skipping GAN training.")
        return [], [], []

    d_losses, g_losses, d_accs = [], [], []
    print("\nStarting GAN training...")

    for epoch in range(epochs):
        d_loss_epoch, g_loss_epoch, d_acc_epoch = 0, 0, 0
        for _ in range(num_batches):
            # Train Discriminator
            disc.trainable = True
            idx = np.random.randint(0, data_for_gan.shape[0], batch_size_gan)
            real_signals = data_for_gan[idx]
            real_labels = np.ones((batch_size_gan, 1)) * 0.9 # Label smoothing

            noise = np.random.normal(0, 1, (batch_size_gan, latent_dim_gan))
            fake_signals = gen.predict(noise, verbose=0)
            fake_labels = np.zeros((batch_size_gan, 1))

            d_loss_real = disc.train_on_batch(real_signals, real_labels)
            d_loss_fake = disc.train_on_batch(fake_signals, fake_labels)
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
            d_loss_epoch += d_loss[0]
            d_acc_epoch += d_loss[1]

            # Train Generator (via combined GAN model)
            disc.trainable = False
            noise = np.random.normal(0, 1, (batch_size_gan, latent_dim_gan))
            misleading_labels = np.ones((batch_size_gan, 1))
            g_loss = combined_gan.train_on_batch(noise, misleading_labels)
            g_loss_epoch += g_loss

        avg_d_loss = d_loss_epoch / num_batches
        avg_d_acc = d_acc_epoch / num_batches
        avg_g_loss = g_loss_epoch / num_batches
        d_losses.append(avg_d_loss)
        g_losses.append(avg_g_loss)
        d_accs.append(avg_d_acc)

        if epoch % (epochs // 10 if epochs >=10 else 1) == 0 or epoch == epochs - 1 :
            print(f"Epoch {epoch}/{epochs}: D_loss={avg_d_loss:.4f}, D_acc={avg_d_acc:.4f}, G_loss={avg_g_loss:.4f}")
            if epoch % (epochs // 2 if epochs >=10 else 1) == 0 or epoch == epochs - 1: # Generate sample less frequently
                sample_noise = np.random.normal(0, 1, (1, latent_dim_gan))
                generated_sample_signal = gen.predict(sample_noise, verbose=0)
                plt.figure(figsize=(10, 3))
                plt.plot(generated_sample_signal[0].flatten())
                plt.title(f'GAN Generated Signal - Epoch {epoch}')
                plt.xlabel('Time Step'); plt.ylabel('Amplitude')
                plt.show()
    print("GAN training finished.")
    return d_losses, g_losses, d_accs

# Train the GAN
# Note: X_train_reshaped is used. Its values are from StandardScaler.
# If GAN performs poorly, consider scaling X_train_reshaped to [-1, 1] for GAN training.
d_losses_hist, g_losses_hist, d_accs_hist = train_gan_loop(
    generator_gan, discriminator_gan, gan_combined_model,
    X_train_reshaped, GAN_TRAIN_EPOCHS, GAN_BATCH_SIZE, LATENT_DIM_GAN
)

if d_losses_hist: # Check if training happened
    plt.figure(figsize=(12, 5))
    plt.plot(d_losses_hist, label='Discriminator Loss')
    plt.plot(g_losses_hist, label='Generator Loss')
    plt.title('GAN Training Losses'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True); plt.show()

    plt.figure(figsize=(12, 5))
    plt.plot(d_accs_hist, label='Discriminator Accuracy', color='green')
    plt.title('Discriminator Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True); plt.show()


# --- Part 5: Classifier Model Definition, Compilation, and Training (using AE Encoder) ---
print("\n--- Part 5: Classifier Model ---")

def build_classifier(encoder_feat_model, num_classes_clf):
    if encoder_feat_model is None:
        raise ValueError("An encoder model must be provided.")

    # Freeze encoder layers
    for layer in encoder_feat_model.layers:
        layer.trainable = False

    input_clf = keras.Input(shape=(SEQUENCE_LENGTH, 1), name='classifier_input')
    encoded_features = encoder_feat_model(input_clf) # Output shape: (None, LATENT_DIM_AE)

    x = layers.Dense(256, activation='relu')(encoded_features)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.4)(x)
    x = layers.Dense(128, activation='relu')(x)
    x = layers.BatchNormalization()(x)
    x = layers.Dropout(0.3)(x)
    x = layers.Dense(64, activation='relu')(x)
    x = layers.Dropout(0.2)(x)
    output_clf = layers.Dense(num_classes_clf, activation='softmax', name='output')(x)
    classifier_model = keras.Model(input_clf, output_clf, name='signal_classifier')
    return classifier_model

num_unique_classes = len(np.unique(y_sequences))

# Compute class weights for imbalanced datasets
class_weights_computed = compute_class_weight(
    'balanced', classes=np.unique(y_train), y=y_train
)
class_weight_map = dict(zip(np.unique(y_train), class_weights_computed))
print(f"\nComputed class weights for classifier: {class_weight_map}")

classifier = build_classifier(encoder, num_classes_clf=num_unique_classes) # `encoder` is from trained AE

classifier.compile(
    optimizer=keras.optimizers.Adam(learning_rate=0.001),
    loss='sparse_categorical_crossentropy', # For integer labels
    metrics=['accuracy']
)

classifier_callbacks = [
    keras.callbacks.EarlyStopping(monitor='val_accuracy', patience=15, restore_best_weights=True, mode='max'),
    keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=7, min_lr=1e-7)
]

print("\nClassifier Architecture (using pre-trained AE encoder):")
classifier.summary()

print("\nStarting Classifier training...")
history_classifier = classifier.fit(
    X_train_reshaped, y_train,
    epochs=CLASSIFIER_TRAIN_EPOCHS,
    batch_size=CLASSIFIER_BATCH_SIZE,
    validation_data=(X_val_reshaped, y_val),
    class_weight=class_weight_map,
    callbacks=classifier_callbacks,
    verbose=1
)
print("Classifier training finished.")

# Plot classifier training history
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history_classifier.history['accuracy'], label='Train Accuracy')
plt.plot(history_classifier.history['val_accuracy'], label='Validation Accuracy')
plt.title('Classifier Accuracy'); plt.xlabel('Epoch'); plt.ylabel('Accuracy'); plt.legend(); plt.grid(True)
plt.subplot(1, 2, 2)
plt.plot(history_classifier.history['loss'], label='Train Loss')
plt.plot(history_classifier.history['val_loss'], label='Validation Loss')
plt.title('Classifier Loss'); plt.xlabel('Epoch'); plt.ylabel('Loss'); plt.legend(); plt.grid(True)
plt.tight_layout()
plt.show()


# --- Part 6: Model Evaluation and Synthetic Data Generation ---
print("\n--- Part 6: Model Evaluation and Synthetic Data Generation ---")

def evaluate_classifier_model(clf_model, X_test_data, y_test_data):
    print("\nEvaluating classifier performance on the test set...")
    y_pred_probs = clf_model.predict(X_test_data)
    y_pred_classes = np.argmax(y_pred_probs, axis=1)

    # Match encoding: {'quiet': 0, 'vehicle': 1, 'human': 2}
    target_names_report = ['quiet', 'vehicle', 'human']

    print("\nClassification Report:")
    print(classification_report(y_test_data, y_pred_classes, target_names=target_names_report, zero_division=0))

    plt.figure(figsize=(8, 6))
    cm = confusion_matrix(y_test_data, y_pred_classes)
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
                xticklabels=target_names_report, yticklabels=target_names_report)
    plt.title('Confusion Matrix - Signal Classification'); plt.ylabel('True Label'); plt.xlabel('Predicted Label')
    plt.show()

evaluate_classifier_model(classifier, X_test_reshaped, y_test)

def generate_and_plot_synthetic_signals(gen_model, latent_dim_val, num_signals=3):
    """
    Generates and plots synthetic signals using the trained GAN generator.
    The current GAN is unconditional (does not generate class-specific data).
    """
    print(f"\nGenerating {num_signals} synthetic signals using the GAN...")
    noise_samples = np.random.normal(0, 1, (num_signals, latent_dim_val))
    synthetic_signals_gen = gen_model.predict(noise_samples, verbose=0)

    # Reshape if necessary for plotting (samples, sequence_length)
    if synthetic_signals_gen.ndim == 3 and synthetic_signals_gen.shape[-1] == 1:
        synthetic_signals_plot = synthetic_signals_gen.reshape(num_signals, -1)
    else:
        synthetic_signals_plot = synthetic_signals_gen

    plt.figure(figsize=(12, num_signals * 2.5))
    for i in range(num_signals):
        plt.subplot(num_signals, 1, i + 1)
        plt.plot(synthetic_signals_plot[i])
        plt.title(f'Synthetic Signal {i+1}')
        plt.ylabel('Amplitude')
        if i == num_signals - 1: plt.xlabel('Time Step')
        plt.grid(True)
    plt.tight_layout()
    plt.show()
    return synthetic_signals_gen

# Generate synthetic signals using the trained GAN's generator
# Note: The GAN is trained on X_train_reshaped. If you intend to use these signals
# for augmenting the classifier's training data, you'd typically generate signals,
# assign labels (if you had a conditional GAN or a heuristic), and add them to the
# classifier's training set, then retrain the classifier. This is not done here.
synthetic_signals_output = generate_and_plot_synthetic_signals(generator_gan, LATENT_DIM_GAN, num_signals=3)

print("\n--- Machine Learning Pipeline Completed ---")
```
